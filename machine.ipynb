```
import gzip
f = gzip.open('/content/train-images-idx3-ubyte.gz','r')

image_size = 28
num_images = 10000

import numpy as np
f.read(100)
buf = f.read(image_size * image_size * num_images)#pixel size of the images
print(len(buf))
data = np.frombuffer(buf, dtype=np.uint8).astype(np.float32)
print(data.shape)
data = data.reshape(num_images, image_size, image_size, 1)#reshaping the pixels to its original size images
print(data[1].shape)
#print(data[1])
```

```
import matplotlib.pyplot as plt
for i in range(0,500):
    image = np.asarray(data[i]).squeeze()
    #image = image/255 # to reduce calculation
    plt.imshow(image)
    plt.show()
```
